```python
x = [1,2,3,4,5]
y= [2,3,7,1,0]


titulo = "Gráfico de Barras"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title (titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)


plt.bar(x,y)
plt.show
```

```python
x1 = [1,3,5,7,9]
y1 = [2,3,7,1,0]

x2 = [2,4,6,8,10]
y2 = [5,1,3,7,4]


titulo = "Gráfico de Barras 2"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title (titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)


plt.bar(x1,y1, label = "Grupo 1")
plt.bar(x2,y2, label = "Grupo 2")
plt.legend()
plt.show
```

```python
x = [1,2,3,4,5]
y= [2,3,7,1,0]


titulo = "scatterplot: Gráfico de Dispersão"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title (titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label="Meus Pontos", color="r")
plt.plot(x,y)
plt.legend()
plt.show
```

```python
x = [1,2,3,4,5]
y= [2,3,7,1,0]


titulo = "scatterplot: Gráfico de Dispersão"
eixox = "Eixo X"
eixoy = "Eixo Y"

#legendas
plt.title (titulo)
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x,y, label="Meus Pontos", color="g", marker=".")
plt.plot(x,y, color="r", linestyle=":")
plt.legend()
plt.show
```

```python
#Crescimento da população brasileira de 1980 - 2016
# Fonte: Datasus
dados = open("populacao-brasileira.csv").readlines()

x =[]
y =[]

for i in range(len(dados)):
    if i != 0:
        linha = dados[i].split(";")
        x.append(int(linha[0]))
        y.append(int(linha[1]))

plt.bar(x,y,color="g")
plt.plot(x,y,color="k", linestyle="--")

plt.title("Crescimento da população brasileira de 1980 - 2016")
plt.xlabel("Ano")
plt.ylabel("População x 100.000.000")
plt.show
```

```python
# Boxplot - Diagrama de caixa
import matplotlib.pyplot as plt
import random

vetor = []

for i in range(10):
    numero_aleatorio = random.randint(0,8)
    vetor.append(numero_aleatorio)
    
plt.boxplot(vetor)
plt.title("Boxplot")
plt.show



```
