# Regressão Linear Simples

```python
#Bibliotecas
import pandas as pd                #carregamento da base
import numpy as np                 #cálculos matemáticos
import matplotlib.pyplot as plt    #visualisação do gráfico
from sklearn.linear_model import LinearRegression  #sklearn para machine learning
from yellowbrick.regressor import ResidualsPlot
```

```python
#Ler o arquivo de ddos e alterações do arquivo
base = pd.read_csv('cars.csv')     #criando (carregando) a base de dados
base = base.drop(['Unnamed: 0'], axis = 1) #apagar o primeiro atributo [nome do atributo:],
                                           #apgar por coluna
```

```python
#criar váriaveis dos eixos e carregar os valores
x = base.iloc[:, 1].values  #.values para transformar em array
y = base.iloc[:, 0].values
```

```python
#Correlação entre os dados
correlacao = np.corrcoef(x, y)
```

```python
#calculo e plotagem da regressão 

x = x.reshape(-1, 1)  #mudou o formato de apresentação do x (de linha para coluna)
modelo = LinearRegression() #instanciando o modelo desta classe
modelo.fit(x, y)      #tentar encaixar estes dados

modelo.intercept_     #visualisar a interseptação
modelo.coef_          #visualisar a inclinação

plt.scatter(x, y, color = 'green')     #plotar o gráfico
plt.plot(x, modelo.predict(x), color = 'yellow')    #traçar a linha da regressão no gráfico
```

```python
#Previsão da velocidade
#considerando como exemplo a distância 22 pés
#modelo.predict(22) - código não funcionou
modelo.intercept_ + modelo.coef_ * 22  #previsão manual
```

```python
#residuais
modelo._residues
```

```python
#plotar residuais
visualizador = ResidualsPlot(modelo)
visualizador.fit(x, y)
visualizador.poof()
```
